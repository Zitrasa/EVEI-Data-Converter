```python
import duckdb
```

```python
database = duckdb.connect("prepared/eve-data.duckdb", read_only = False)
```

```python
database.execute(
    "CREATE TABLE items AS SELECT * FROM read_csv_auto("
        "'prepared/item_data.csv',"
        "delim = ',',"
        "header = True,"
        "columns = {"
            "'id': 'UINTEGER',"
            "'blueprint_id': 'UINTEGER',"
            "'volume': 'DOUBLE',"
            "'name': 'VARCHAR'"
        "}"
    ")"
)
```

```python
database.execute(
    "CREATE TABLE blueprints AS SELECT * FROM read_csv_auto("
        "'prepared/blueprint_data.csv',"
        "delim = ',',"
        "header = True,"
        "columns = {"
            "'id': 'UINTEGER',"
            "'materials_id': 'UINTEGER[]',"
            "'quantities': 'UINTEGER[]',"
            "'manufacture_time': 'UINTEGER',"
            "'time_research_time': 'UINTEGER',"
            "'material_research_time': 'UINTEGER',"
            "'copy_time': 'UINTEGER',"
            "'copied_max_runs': 'UINTEGER',"
            "'product_id': 'UINTEGER',"
            "'product_quantity': 'UINTEGER',"
            "'materials_count': 'UINTEGER',"
            "'name': 'VARCHAR'"
        "}"
    ")"
)
```

```python
database.execute(
    "CREATE TABLE inventions AS SELECT * FROM read_csv_auto("
        "'prepared/invention_data.csv',"
        "delim = ',',"
        "header = True,"
        "columns = {"
            "'origin_id': 'UINTEGER',"
            "'materials_id': 'UINTEGER[]',"
            "'quantities': 'UINTEGER[]',"
            "'invention_time': 'UINTEGER',"
            "'invented_id': 'UINTEGER',"
            "'invented_runs': 'UINTEGER',"
            "'materials_count': 'UINTEGER'"
        "}"
    ")"
)
```

```python
database.execute("SELECT * FROM items").fetchdf()
```

```python
database.execute("SELECT * FROM blueprints").fetchdf()
```

```python
database.execute("SELECT * FROM inventions").fetchdf()
```

```python
database.close()
```
